```python
import pandas as pd
from rdkit import Chem

def count_non_h_atoms(inchi):
    mol = Chem.MolFromInchi(inchi)
    if mol is not None:
        return mol.GetNumAtoms()
    else:
        return None
```

```python
m = Chem.MolFromSmiles("CCC")
m.GetNumAtoms()
```

```python
df = pd.read_csv("./data/InChI_all/training_data_all.csv")
df["n_non_H_atoms"] = df.InChI.apply(count_non_h_atoms)
```
